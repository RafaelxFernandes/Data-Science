```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/alura-clustering-validation/base-de-dados/CC%20GENERAL.csv")
df.head()
```

```python
df.drop(columns = ["CUST_ID", "TENURE"], inplace = True)
df.head()
```

```python
df.isna().sum()
```

```python
df.fillna(df.median(), inplace = True)
df.isna().sum()
```

```python
from sklearn.preprocessing import Normalizer

values = Normalizer().fit_transform(df.values)
values
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 5, n_init = 10, max_iter = 300)
y_pred = kmeans.fit_predict(values)
```

```python
from sklearn import metrics

labels = kmeans.labels_
silhouette = metrics.silhouette_score(values, labels, metric = "euclidean")
silhouette
```

Varia entre -1 e 1, e quanto mais próximo de 1, mais compactos e separados estão os clusters. Um valor positivo já é um indicativo de boa criação dos clusters. É o índice mais popular e mais utilizado.

```python
dbs = metrics.davies_bouldin_score(values, labels)
dbs
```

Quanto mais próximo de 0, melhor, e quanto mais próximo de 1, pior. Pegando o pior caso, e obtendo um bom resultado, podemos dizer que temos então uma clusterização boa.

```python
calinski = metrics.calinski_harabasz_score(values, labels)
calinski
```

Quanto mais alto, melhor.

```python
def clustering(n_clusters, dataset):
    
    kmeans = KMeans(n_clusters, n_init = 10, max_iter = 300)
    labels = kmeans.fit_predict(dataset)
    
    silhouette = metrics.silhouette_score(dataset, labels, metric = "euclidean")
    
    dbs = metrics.davies_bouldin_score(dataset, labels)
    
    calinski = metrics.calinski_harabasz_score(dataset, labels)
    
    return silhouette, dbs, calinski
```

```python
s1, dbs1, calinski1 = clustering(3, values)
print(s1, dbs1, calinski1)
```

```python
s2, dbs2, calinski2 = clustering(5, values)
print(s2, dbs2, calinski2)
```

```python
s3, dbs3, calinski3 = clustering(10, values)
print(s3, dbs3, calinski3)
```

```python
import numpy as np

random_data = np.random.rand(8950, 16)
s_random, dbs_random, calinski_random = clustering(5, random_data)
print(s_random, dbs_random, calinski_random)
```

```python
set1, set2, set3 = np.array_split(values, 3)
s1, dbs1, c1 = clustering(5, set1)
s2, dbs2, c2 = clustering(5, set2)
s3, dbs3, c3 = clustering(5, set3)
```

```python
print(s1, dbs1, c1)
print(s2, dbs2, c2)
print(s3, dbs3, c3)
```

Valores próximos são evidências de clusters estáveis.
