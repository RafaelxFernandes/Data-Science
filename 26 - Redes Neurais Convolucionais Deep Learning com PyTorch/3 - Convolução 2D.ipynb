```python
from skimage import io, color, transform, data
from scipy.signal import convolve
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = data.brick()
plt.imshow(img, cmap='Greys')
```

```python
def show(valores, title):
    
    plt.figure(figsize=(len(valores), len(valores)))
    plt.imshow(valores, cmap='gray')
    
    for i, line in enumerate(valores):
        for j, col in enumerate(line):
            plt.text(j, i, '{:.0f}'.format(col), fontsize=16, color='red', ha='center', va='center')
            
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.savefig(title+'.png', format='png', dpi=100, bbox_inches='tight')
```

```python
kernel_vertical = [
    [-1, 0, 1],
    [-1, 0, 1],
    [-1, 0, 1]
]

show(kernel_vertical, 'Kernel vertical')
```

```python
kernel_horizontal = [
    [-1, -1, -1],
    [0, 0, 0],
    [1, 1, 1]
]

show(kernel_horizontal, 'Kernel horizontal')
```

```python
mapa_caracteristicas = convolve(img, kernel_vertical, mode='valid')

plt.imshow(img, cmap='Greys')

plt.figure()
plt.imshow(mapa_caracteristicas, cmap='Greys')
```

```python
mapa_caracteristicas = convolve(img, kernel_horizontal, mode='valid')

plt.imshow(img, cmap='Greys')

plt.figure()
plt.imshow(mapa_caracteristicas, cmap='Greys')
```
