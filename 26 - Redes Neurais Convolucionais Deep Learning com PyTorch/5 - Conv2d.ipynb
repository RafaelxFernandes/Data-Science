```python
import torch
from torch import nn

import numpy as np
import matplotlib.pyplot as plt
from skimage import data
```

```python
pb = data.brick()
rgb = data.astronaut()

plt.imshow(pb, cmap='Greys')
plt.figure()
plt.imshow(rgb)
```

```python
conv = nn.Conv2d(1, 16, kernel_size=3)

pb_tensor = torch.Tensor(pb)
pb_tensor = pb_tensor.view(1, 1, pb_tensor.size(0), pb_tensor.size(1))

mapa_ativacao = conv(pb_tensor)
```

```python
conv = nn.Conv2d(3, 16, kernel_size=3)

rgb_tensor = torch.Tensor(rgb)
rgb_tensor = rgb_tensor.permute(2, 0, 1).unsqueeze(0)

mapa_ativacao = conv(rgb_tensor)
```
