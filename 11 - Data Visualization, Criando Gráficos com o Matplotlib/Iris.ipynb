```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("iris.csv")
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.scatter(df["comprimento_sépala"], df["largura_sépala"])
eixo.set_title("Gráfico de dispersão", fontsize = 25, pad = 15)
eixo.set_xlabel("Comprimento da sépala", fontsize = 15)
eixo.set_ylabel("Largura da sépala", fontsize = 15)
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

eixo.scatter(df["comprimento_sépala"], df["largura_sépala"])
eixo.set_title("Gráfico de dispersão", fontsize = 25, pad = 15)
eixo.set_xlabel("Comprimento da sépala", fontsize = 15)
eixo.set_ylabel("Largura da sépala", fontsize = 15)
eixo.tick_params(labelsize = 15)
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {
    "Iris-setosa": "r",
    "Iris-versicolor": "b",
    "Iris-virginica": "g"
}

for especie in df["espécie"].unique():
    df_temp = df[df["espécie"] == especie]
    eixo.scatter(df_temp["comprimento_sépala"], df_temp["largura_sépala"], color = cores[especie])

eixo.set_title("Gráfico de dispersão", fontsize = 25, pad = 15)
eixo.set_xlabel("Comprimento da sépala", fontsize = 15)
eixo.set_ylabel("Largura da sépala", fontsize = 15)
eixo.tick_params(labelsize = 15)
eixo.legend(cores, fontsize = 20)
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {
    "Iris-setosa": "r",
    "Iris-versicolor": "b",
    "Iris-virginica": "g"
}

marcadores = {
    "Iris-setosa": "x",
    "Iris-versicolor": "o",
    "Iris-virginica": "v"
}

for especie in df["espécie"].unique():
    df_temp = df[df["espécie"] == especie]
    eixo.scatter(df_temp["comprimento_sépala"], df_temp["largura_sépala"], 
                 color = cores[especie], marker = marcadores[especie])

eixo.set_title("Gráfico de dispersão", fontsize = 25, pad = 15)
eixo.set_xlabel("Comprimento da sépala", fontsize = 15)
eixo.set_ylabel("Largura da sépala", fontsize = 15)
eixo.tick_params(labelsize = 15)
eixo.legend(cores, fontsize = 20)
```

```python
fig = plt.figure(figsize = (15, 8))
eixo = fig.add_axes([0, 0, 1, 1])

cores = {
    "Iris-setosa": "r",
    "Iris-versicolor": "b",
    "Iris-virginica": "g"
}

marcadores = {
    "Iris-setosa": "x",
    "Iris-versicolor": "o",
    "Iris-virginica": "v"
}

for especie in df["espécie"].unique():
    df_temp = df[df["espécie"] == especie]
    eixo.scatter(df_temp["comprimento_sépala"], df_temp["largura_sépala"], 
                 color = cores[especie], marker = marcadores[especie],
                s = 100)

eixo.set_title("Gráfico de dispersão", fontsize = 25, pad = 15)
eixo.set_xlabel("Comprimento da sépala", fontsize = 15)
eixo.set_ylabel("Largura da sépala", fontsize = 15)
eixo.tick_params(labelsize = 15)
eixo.legend(cores, fontsize = 20)
```

```python
fig = plt.figure(figsize = (5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

# Média e desvio padrão
mu, sigma = df["comprimento_pétala"].mean(), df["comprimento_pétala"].std()

eixo.hist(df["comprimento_pétala"], bins = 20)
eixo.set_title("Histograma", fontsize = 15, pad = 10)
eixo.set_xlabel("Comprimento da pétala", fontsize = 12)
eixo.grid(True)

eixo.annotate("$mu = {0:.2f}$\n$sigma = {1:.2f}$".format(mu, sigma),
             xy = (4.5, 20), fontsize = 20)
```

```python
fig = plt.figure(figsize = (5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

# Média e desvio padrão
mu, sigma = df["comprimento_pétala"].mean(), df["comprimento_pétala"].std()

eixo.hist(df["comprimento_pétala"], bins = 20)
eixo.set_title("Histograma", fontsize = 15, pad = 10)
eixo.set_xlabel("Comprimento da pétala", fontsize = 12)
eixo.grid(True)

eixo.annotate("$mu = {0:.2f}$\n$sigma = {1:.2f}$".format(mu, sigma),
             xy = (4.5, 20), fontsize = 20)

eixo.axvline(mu, color = "k", linestyle = "--")
eixo.annotate("Média", xy = (mu - 1.3, 28), fontsize = 20)
```

```python
fig = plt.figure(figsize = (5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

# Média e desvio padrão
mu, sigma = df["comprimento_pétala"].mean(), df["comprimento_pétala"].std()

eixo.hist(df["comprimento_pétala"], bins = 20)
eixo.set_title("Histograma", fontsize = 15, pad = 10)
eixo.set_xlabel("Comprimento da pétala", fontsize = 12)
eixo.grid(True)

eixo.annotate("$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu, sigma),
             xy = (4.5, 20), fontsize = 20)

eixo.axvline(mu, color = "k", linestyle = "--")
eixo.annotate("Média", xy = (mu - 1.3, 28), fontsize = 20)

eixo.axvline(df["comprimento_pétala"].median(), color = "g", linestyle = "--")
eixo.annotate("Mediana", xy = (df["comprimento_pétala"].median(), 31), fontsize = 20)
```

```python
fig = plt.figure(figsize = (5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

df_versicolor = df[df["espécie"] == "Iris-versicolor"]

# Média e desvio padrão
mu, sigma = df_versicolor["comprimento_pétala"].mean(), df_versicolor["comprimento_pétala"].std()

eixo.hist(df_versicolor["comprimento_pétala"], bins = 20)
eixo.set_title("Iris Versicolor", fontsize = 15, pad = 10)
eixo.set_xlabel("Comprimento da pétala", fontsize = 12)
eixo.grid(True)

eixo.annotate("$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu, sigma),
             xy = (4.6, 6), fontsize = 20)

eixo.axvline(mu, color = "k", linestyle = "--")
eixo.annotate("Média", xy = (mu - 0.5, 5.5), fontsize = 20)

eixo.axvline(df_versicolor["comprimento_pétala"].median(), color = "g", linestyle = "--")
eixo.annotate("Mediana", xy = (df_versicolor["comprimento_pétala"].median() - 0.7, 6.5), fontsize = 20, color = "g")

fig.savefig("histograma_iris_versicolor.png", bbox_inches = "tight")
```

```python
fig = plt.figure(figsize = (5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

df_setosa = df[df["espécie"] == "Iris-setosa"]

# Média e desvio padrão
mu, sigma = df_setosa["comprimento_pétala"].mean(), df_setosa["comprimento_pétala"].std()

eixo.hist(df_setosa["comprimento_pétala"], bins = 20)
eixo.set_title("Iris Setosa", fontsize = 15, pad = 10)
eixo.set_xlabel("Comprimento da pétala", fontsize = 12)
eixo.grid(True)

eixo.annotate("$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu, sigma),
             xy = (1.7, 7), fontsize = 20)

eixo.axvline(mu, color = "k", linestyle = "--")
eixo.annotate("Média", xy = (mu - 0.2, 13.5), fontsize = 20)

eixo.axvline(df_setosa["comprimento_pétala"].median(), color = "g", linestyle = "--")
eixo.annotate("Mediana", xy = (df_setosa["comprimento_pétala"].median() + 0.05, 12), fontsize = 20, color = "g")

fig.savefig("histograma_iris_setosa.png", bbox_inches = "tight")
```

```python
fig = plt.figure(figsize = (5, 4))
eixo = fig.add_axes([0, 0, 1, 1])

df_virginica = df[df["espécie"] == "Iris-virginica"]

# Média e desvio padrão
mu, sigma = df_virginica["comprimento_pétala"].mean(), df_virginica["comprimento_pétala"].std()

eixo.hist(df_virginica["comprimento_pétala"], bins = 20)
eixo.set_title("Iris Virginica", fontsize = 15, pad = 10)
eixo.set_xlabel("Comprimento da pétala", fontsize = 12)
eixo.grid(True)

eixo.annotate("$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu, sigma),
             xy = (6.3, 7), fontsize = 20)

eixo.axvline(mu, color = "k", linestyle = "--")
eixo.annotate("Média", xy = (mu, 8), fontsize = 20)

eixo.axvline(df_virginica["comprimento_pétala"].median(), color = "g", linestyle = "--")
eixo.annotate("Mediana", xy = (df_virginica["comprimento_pétala"].median(), 7), fontsize = 20, color = "g")

fig.savefig("histograma_iris_virginica.png", bbox_inches = "tight")
```

```python
from PIL import Image

largura, altura = Image.open("histograma_iris_virginica.png").size
combinada = Image.new("RGB", (3 * largura, altura))
intervalo = 0

for imagem in map(Image.open, ["histograma_iris_versicolor.png", "histograma_iris_setosa.png", 
                               "histograma_iris_virginica.png"]):

    combinada.paste(imagem, (intervalo, 0))
    intervalo += largura
    
combinada.save("combinada.png")
```
